## Пункт 3.1

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

with open("data1.txt", 'r') as f:
    x = f.readline().replace('\n', '').split(' ')[:-1]
    x = list(map(int, x))
    y = f.readline().replace('\n', '').split(' ')[:-1]
    y = list(map(int, y))

plt.figure(figsize=(30,5))
plt.plot(x, y)
plt.show()
```
